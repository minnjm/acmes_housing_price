## Data Exploration

** Import training and test datasets. Check the sizes of both.**

```python
df = pd.read_csv("../data/train.csv", index_col=0)
train = df.iloc[:, :-1]
y = df.iloc[:, -1]
test = pd.read_csv("../data/test.csv", index_col=0)

print("Ames housing training dataset has {} data points with {} variables each.".format(train.shape[0], train.shape[1]))
print("Ames housing test dataset has {} data points with {} variables each.".format(test.shape[0], test.shape[1]))
```

** Note: Training and test datasets are about the same size. Need to be mindful of following two implications: **
- Some data points from test dataset may spread outside training dataset occupying space, so make the price prediction for these data points less accurate.
- Some values in categorical features of test dataset may not be presented in training dataset, so the model may break during the prediction of house prices for test dataset. One work-around is concatenate train and test datasets as below.

```python
train_test = pd.concat([train, test])
```

** Note: Some features are missing values for almost all the houses, for example PoolQC, Fence, etc..  Data description file provided reveals some insights: For example **

PoolQC: Pool quality	
    - Ex	Excellent
    - Gd	Good
    - TA	Average/Typical
    - Fa	Fair
    - NA	No Pool
       
** The value is not missing, is just simply 'NA' and pandas inteprets it as 'NaN'.**
** The "missing" values should not be filled using convensional method, ie. with most frequent value for categorical features. I impute this type of 'NaN' with unique value "MISSING". This is applicable to following features in valcolumns list: **

```python
valcolumns = ['Alley', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'FireplaceQu',
              'GarageQual', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond', 'PoolQC',
              'Fence', 'MiscFeature']
```
